```python
import requests
import re
import numpy as np
from bs4 import BeautifulSoup
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
TITLES = []
RATING = []
AREA = []
DISHES = []
APPROX = []
DISC = []

city = ['ahmedabad', 'mumbai', 'pune', 'bengaluru']
for c in city:
    for i in range(1,16):
        url = f"https://www.eazydiner.com/restaurants?location={c}&page={i}"
        requests.get(url)
        page = requests.get(url)
        soup = BeautifulSoup(page.text)
        title = soup.find_all('a', class_='ellipsis listing_res_name__uVIN8')
        for i in title:
            TITLES.append(i.text)
        rating = soup.find_all('tspan')
        for i in rating:
            j = i.text
            if j != '%':
                if j:
                    RATING.append(j)
                else:
                    RATING.append(np.nan)
        area = soup.find_all('div', class_='font-12 ellipsis flex listing_res_address__LuVOm')
        for i in area:
            AREA.append(i.text)
        dishes = soup.find_all('div', class_='flex align-v-center ellipsis listing_multi_cuisines__zLLXK')
        for i in dishes:
            if i:
                DISHES.append(i.text)
            else:
                DISHES.append(np.nan)
        txt = []
        approx = soup.find_all('div', class_='flex align-v-center')
        for i in approx:
            txt.append(i.text)
        for i in txt:
            if re.match("₹ +\d+ [A-Za-z].+", i):
                APPROX.append(i)
            else:
                pass
        discount = soup.find_all('div', class_='ellipsis listing_offer_text__LXKw1')
        for i in discount:
            DISC.append(i.text)

print(len(TITLES))
print(len(RATING))
print(len(AREA))
print(len(DISHES))
print(len(APPROX))
print(len(DISC))
```

```python
data={
    'NAME': TITLES,
    'RATING': RATING,
    'AREA': AREA,
    'DISHES': DISHES,
    'APPROX': APPROX,
    'DISC': DISC
}
df = pd.DataFrame(data)
df
```

```python
df.to_csv('restaurant_data.csv', index=False)
```
